```python
def process_and_pivot_p(df, id_vars, chunk_size=1000):
    
    processed_chunks = []
    
    total_chunks = (len(df) + chunk_size - 1) // chunk_size
    
    for chunk_num, start_idx in enumerate(range(0, len(df), chunk_size), 1):
        
        print(f"Processing chunk {chunk_num} of {total_chunks} ({(chunk_num/total_chunks*100):.1f}%)")
        
        chunk = df.iloc[start_idx:start_idx + chunk_size].copy()
        
        chunk_long = chunk.melt(
            id_vars=id_vars,
            var_name="event_id", 
            value_name="event_value"
        )
        
        chunk_long = chunk_long.rename(
            columns={"CNTSTUID": "id"}
        )
        
        conditions = [
            chunk_long["event_id"].str.endswith("TT"),
            chunk_long["event_id"].str.endswith("F"),
            chunk_long["event_id"].str.endswith("A"),
            chunk_long["event_id"].str.endswith("V")
        ]
        
        choices = [
            "rt",
            "time_to_first_action",
            "num_actions",
            "num_visits"
        ]
        
        chunk_long["item"] = np.where(
            chunk_long["event_id"].str.endswith("TT"), 
            chunk_long["event_id"].str[:-2], 
            chunk_long["event_id"].str[:-1]
        )
        
        chunk_long["event_type"] = np.select(conditions, choices, default="unknown")
        chunk_long = chunk_long.dropna(subset=["event_value"])
        
        processed_chunks.append(chunk_long)
        
        del chunk_long
    
    print("Combining chunks...")
    df_long = pd.concat(processed_chunks, ignore_index=True)

    return df_long
```

```python
def process_and_pivot_by_cnt(df):
    
    unique_cnts = df['CNT'].unique()
    print(f"Processing {len(unique_cnts)} unique countries...")
    
    pivoted_chunks = []
    
    for i, cnt in enumerate(unique_cnts, 1):
        print(f"Processing country {i} of {len(unique_cnts)}: {cnt}")
        
        cnt_data = df[df['CNT'] == cnt]
        
        index_cols = ['id', 'CNT', 'CNTSCHID', 'item', 'event_type']
        
        try:
            pivoted_chunk = cnt_data.pivot(
                index=['id', 'CNT', 'CNTSCHID', 'item'],
                columns='event_type',
                values='event_value'
            ).reset_index()
            
            pivoted_chunks.append(pivoted_chunk)
            
            del cnt_data
            del pivoted_chunk
            
        except Exception as e:
            print(f"Error processing {cnt}: {str(e)}")
            continue
    
    print("Combining final results...")
    df_wide = pd.concat(pivoted_chunks, ignore_index=True)
    
    print("Done!")
    return df_wide
```
